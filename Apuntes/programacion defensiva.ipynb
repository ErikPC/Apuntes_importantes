# Programacion defensiva

## Proteccion del programa contra entradas no válidas

1. Comprueba que los valores de todos los datos de fuentes externas.
2. Comprobrar los valores de todos los párametros de entreada de rutinas.
3. Decide cómo manejar entradas incorrectas.

La mejor manera de "defensive coding" es **no insertar errores en primer lugar. Usar un diseño iterativo, ezcribir pseudocodigo antes del código, escribir casos test antes de escribir el codigo y realizar inspecciones de diseño de bajo nivel.

## Assertions

- Es el codigo que se usa durante el desarrollo, permite que un programa se compruebe a sí mismo a medida que se ejecuta.

- Una assertion generalmente tomas dos rgumentos: una expresion booleana que describe la suposicion que se supoque que es verdadera, y un mensaje para mostrar si no lo es.

- Normalmente se usan durante el desarrollo y el mantenimiento.

### Guia para usar Asserions

- El manejo de errores generalmente verifica si hay datos de entrada incorrectors; las assertions comprueban si hay errores en el codigo.

- La accion correcta es cambiar el código fuente del programa.

- El compilador eliminara el código cuando desactive las assertions.

#### Utilice las aserciones para documentar y verificar las condiciones previas y posteriores.

"Design by contract"

- Las precondiciones son las propiedades que el código de cliente de una rutina o clase promete que serán verdaderas antes de llamar a la rutina o crear una instancia del objeto.

- Las postcondiciones son las propiedades que la rutina o clase promete que serán verdaderas cuando concluya su ejecución.

## Tecnicas de manejo de errores

Dependiendo de unas circunstancias especificas.

- desea devolver un valor neutral
- sustituir el siguiente dato válido
- devolver la misma respuesta que la vez anterior
- sustituir el valor legal más cercano
- registrar un mensaje de advertencia en un archivo()
- devolver un código de error
- llamar a una rutina u objeto de procesamiento de errores
- mostrar un mensaje de error
- o apagar
- o es posible que desee utilizar una combinacion de estas respuestas.

### Devolviendo valor neutral

Continua operando y simplemente devuelva un valor que se sabe que es inofensivo, evolviendo los siguientes datos válidos.

### Devolviendo un codigo de error

Sólo ciertas partes de un sistema controlarán los errores, otra rutina más arriba en la jerarquia de llamadas controlará el error. 

### Llamar a una rutina/objeto de procesamiento de errores

Centralizar el control de errores en una rutina global de control de errores o en un objeto de control de errores.

### Muestra un mensaje de error dondequiera que se encuentre el error

Los atacantes a veces usan mensajes de error para descubrir cómo atacar un sistema.

### Robustez vs correccion

El estilo de procesamiento de errores que es más apropiado depende del tipo de software en el que se produce el error.

- Correccion significa nunca devolver un valor inexacto.
- Robustez significa , siempre tratar de hacer halgo que permita que el software siga funcionando
